```python
%matplotlib inline

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
A = np.matrix([[0.8, 0.3], [0.2, 0.7]])
A
```

```python
def markov_multiple(vec, vec_name, stop, ax1, ax2, y_limits, cmap):
    vec_copy = vec.copy()
    vec_list = np.zeros((stop, 2))
    labels = []
    for i in range(1, stop+1):
        vec_copy = np.asarray(np.dot(A, vec_copy)).ravel()
        print('{}: {}'.format(i, vec_copy))
        labels.append(i)
        vec_list[i - 1] = vec_copy
    

    ax1.plot(vec_list[:, 0], vec_list[:, 1], '-o', label=vec_name)
    ax1.legend()
    
    def annotate(label, vec, ax, v_shift):
        ax.annotate(label, xy=(vec[0], vec[1]), 
                         xytext=(vec[0], vec[1] + v_shift),
                         arrowprops=dict(facecolor='black',
                                        width=1, headwidth=5, headlength=7))
    
    shift = 0.5
    max_y = vec_list[:, 1].max()
    min_y = vec_list[:, 1].min()
    y_upper = max_y + shift
    y_lower = min_y - shift

    if y_upper > y_limits.get('upper', -np.inf):
        y_limits['upper'] = y_upper
    if y_lower < y_limits.get('lower', np.inf):
        y_limits['lower'] = y_lower    

    ax1.set(ylim = (y_limits['lower'], y_limits['upper']))
    for i, (label, vec) in enumerate(zip(labels, vec_list)):
        if i%2 == 0:
            annotate(label, vec, ax1, shift)
        else:
            annotate(label, vec, ax1, -shift)
    
    pd.Series(vec_list[:, 0], index=labels).plot(ax=ax2, label=vec_name+'_x',
                                                 color=mpl.cm.get_cmap(cmap)(0.9))
    pd.Series(vec_list[:, 1], index=labels).plot(ax=ax2, label=vec_name+'_y',
                                                 color=mpl.cm.get_cmap(cmap)(0.4))
    ax2.legend()
    

fig, ax = plt.subplots(2, 1)
y_limits = {}
u = np.array([1, 0])
markov_multiple(u, 'u', 7, ax[0], ax[1], y_limits, 'Blues')
v = np.array([0, 1])
markov_multiple(v, 'v', 7, ax[0], ax[1], y_limits, 'Reds')
```
